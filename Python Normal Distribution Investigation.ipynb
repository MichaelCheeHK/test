# Explore the scipy.norm functions

```python
def draw_z_score(x, cond, mu, sigma, title):
    y = norm.pdf(x, mu, sigma)
    z = x[cond]
    plt.plot(x, y)
    plt.fill_between(z, 0, norm.pdf(z, mu, sigma))
    plt.title(title)
    plt.show()
```

### 1. norm.ppf

```python
from scipy.stats import norm
import numpy as np
# norm.ppf() takes a percentage and returns a standard deviation multiplier for what value that percentage occurs at.
# The function returns the standard deviation multiplier given the area under the normal curve
# But the result is one-tail, therefore, the absolute number of 0.01 or 0.99 would equal to one-tail z-score of 98%
round(norm.ppf(0.01, loc=0, scale=1),3)
```

```python
# if I draw the standard normal curve, ppf(0.01) would return the value in x-axis when the left highlighted areas is 1%
x = np.arange(-5,5,0.001)
z0 = -2.36
draw_z_score(x, x<z0, 0, 1, 'z<-2.36')
```

```python
round(norm.ppf(0.99, loc=0, scale=1),3)
```

```python
x = np.arange(-5,5,0.001)
z0 = 2.36
draw_z_score(x, x<z0, 0, 1, 'z<2.36')
```

### 2. norm.pdf

```python
# In order to explian pdf function, let's first generate a dataset with 100 observations
# The numpy.linspace() function returns number spaces evenly w.r.t interval.
# Therefore, x would be an array with 100 values from -2.326 to 2.326 with evenly interval
x = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)
```

```python
# norm.pdf is the probability densify function. Therefore, norm.pdf is a measure of how dense the distribution is around 
# that value. 
y=norm.pdf(x, loc=0, scale=1)
```

```python
# If I plot x, y it would be a normal distribution curve with cutover at -2.326 and 2.326
import matplotlib.pyplot as plt
fig, ax=plt.subplots(1,1)
ax.plot(x, y, 'r-', label='norm pdf')
```

### 3. norm.rvs

```python
# norm.rvs could generate random numbers that follow normal distribution. 
# therefore, it could mimic the sample mean of taking random sample 1000 times. 
# Assume the population mean is 0 and variance is 1
rv=norm.rvs(loc=0, scale=1, size=1000)
```

```python
plt.hist(rv, bins=50)
plt.show()
```

### 4. norm.cdf

```python
# cdf would calculate the probability of x<=given value
print(str(round(norm.cdf(-2.326,loc=0, scale=1)*100,2))+'%')
```

```python
print(str(round(norm.cdf(2.326,loc=0, scale=1)*100,2))+'%')
```

### 5. norm.sf

```python
# norm.sf could convert z-score to p-value
print(str(round(norm.sf(abs(2.326),loc=0, scale=1)*100,2))+'%') #one tail
```

```python
print(str(round(norm.sf(abs(2.326),loc=0, scale=1)*2*100,2))+'%') #two tail
```
